```python
import tensorflow as tf
import pandas as pd
import numpy as np
import time
import scipy
from scipy import linalg

data_frame=pd.read_csv("heart.csv")
#print(data_frame[data_frame.columns[0]])
n_epochs=1000
batch_size = 350

w, h = 462,9;
norms = [[0 for x in range(w)] for y in range(h)] 

X = tf.placeholder(tf.float32, [None,9])# 0 -9 digits recognition = 10 classes
#print(sess.run(X))
Y = tf.placeholder(tf.float32, [None,])


#data = data_frame[data_frame.columns[0]]
init = tf.global_variables_initializer()
sess = tf.Session()
sess.run(init)


#norm = tf.placeholder(tf.float32,[350,9])
for i in range(0,9):
    data = data_frame[data_frame.columns[i]]
    if(i!=4):
        data_norm = [number/scipy.linalg.norm(data) for number in data]
        norms[i]=data_norm
    else:
        for j in range(len(data)):
            if(data[j]=="Present"):
                data[j]=1.0
            else:
                data[j]=0.0
        data_norm = [number for number in data]
        norms[i]=data_norm  
        
        
#print(norms)    

X_train = np.asarray(norms)
X_train = X_train[0:9,0:350]
X_train = map(np.float32, X_train)
X_train = np.transpose(X_train)

Y_train =[number for number in data_frame[data_frame.columns[9]]][:350]
Y_train = map(np.float32, Y_train)
Y_train = np.transpose(Y_train)


X_test = np.asarray(norms)
X_test = X_test[0:9,350:462]
X_test = map(np.float32, X_test)
X_test = np.transpose(X_test)

Y_test =[number for number in data_frame[data_frame.columns[9]]][350:462]
Y_test = map(np.float32, Y_test)
Y_test = np.transpose(Y_test)


# Step 3
W = tf.Variable(tf.random_normal([9, 1], stddev=0.1, mean = 0.0))
b = tf.zeros([350,1], tf.float32)

#X_train = map(np.float32, X_train)

#X_train = X_train.todense()
#X_train = tf.convert_to_tensor(X_train)
#X_train = tf.argmax(X_train, axis=1)
# Step 4

logits = tf.matmul(X_train, W) + b
logits = tf.transpose(logits)

#logits=tf.reshape(logits,(1,350))
hypothesis = tf.nn.softmax(logits)
#print(hypothesis)
entropy=tf.nn.softmax_cross_entropy_with_logits(logits=logits, labels=Y, name='loss')
loss = tf.reduce_mean(entropy)

# Step 5:

optimizer = tf.train.AdamOptimizer(learning_rate=0.1).minimize(loss)

# Step 6:
Y_train = np.transpose(Y_train)
print(len(Y_train))
# Calculate accuracy

with tf.Session() as sess:
    start_time = time.time()
    sess.run(tf.global_variables_initializer())	
    n_batches = int(350/batch_size)
    for i in range(n_epochs): 
        total_loss = 0

        for _ in range(n_batches):
            
            # Run the optimizer and loss, store the loss value as loss_batch
            
            loss_batch, _ = sess.run([loss, optimizer], feed_dict={X: X_train, Y: Y_train})
            total_loss += loss_batch
            #avg_cost += loss_batch / total_batch
        print('Average loss epoch {0}: {1}'.format(i, total_loss/n_batches))

    print('Total time: {0} seconds'.format(time.time() - start_time))
    
    is_correct = tf.equal(tf.argmax(hypothesis,1),tf.argmax(Y,0))
    accuracy = tf.reduce_mean(tf.cast(is_correct, tf.float32))
    # test the model
    preds = tf.nn.softmax(logits)
    correct_preds = tf.equal(tf.argmax(preds, 1), tf.argmax(Y, 0))
    accuracy = tf.reduce_sum(tf.cast(correct_preds, tf.float32)) 
    #print('Accuracy {0}'.sess.run(accuracy))
    print(Y_test)
    n_batches = int(112/4)
    total_correct_preds = 0
    for i in range(n_batches):
        accuracy_batch = sess.run([accuracy], feed_dict={X: X_test, Y:Y_test}) 
        total_correct_preds += sum(accuracy_batch)
    
    print(total_correct_preds)
    print('Accuracy {0}'.total_correct_preds/112)
    
```
